```python
names_1 = pd.read_csv('names/yob2017.txt', names = ['Name', 'Gender', 'Count'])
names_1['Year'] = 2017
names_2 = pd.read_csv('names/yob2016.txt', names = ['Name', 'Gender', 'Count'])
names_2['Year'] = 2016
names_3 = pd.read_csv('names/yob2015.txt', names = ['Name', 'Gender', 'Count'])
names_3['Year'] = 2015
```

```python
names_res = pd.concat([names_1, names_2, names_3])
names_res
```

```python
names_1[names_1['Name'] == 'Aadam']
```

```python
names_2[names_2['Name'] == 'Aadam']
```

```python
names_3[names_3['Name'] == 'Aadam']
```

```python
names_res[names_res['Name'] == 'Aadam']
```

```python
names_res.groupby(['Name']).sum().head(10)
```

```python
names_res.groupby(['Name', 'Gender']).sum().head(100)
```

```python
names_res.groupby(['Name', 'Gender']).agg(['sum', 'max'])
```

```python
names_res['Name_lenght'] = names_res.apply(lambda x: len(x['Name']), axis=1)
names_res
```

```python
res = names_res.groupby(['Gender', 'Year']).mean()[['Name_lenght']]
res
```
